```
materias = ("Matemática" , "Português" , "Ciências")
nome = ("Nome do aluno:")
idade = int (input("Idade do aluno:"))
notas = input("Digite as notas separadas por vígulas:")
lista_notas = [float(n) for n in notas.split(",")]
notas_unicas =  set(lista_notas)
def calcular_media(notas):
  return sum(notas) / len(notas)
media = calcular_media(notas_unicas)
aluno = {
    "nome": nome,
    "idade": idade,
    "notas": list(notas_unicas),
    "media": media
}
if media >= 7:
  situacao = "aprovado!"
else:
  situacao = "reprovado!"
with open("aluno.txt", "w") as arquivo:
  arquivo.write(f"Aluno: {nome}\n")
  arquivo.write(f"Idade: {idade}\n")
  arquivo.write(f"Notas: {aluno ['notas']}\n")
  arquivo.write(f"Média: {media:.2f}\n")
  arquivo.write(f"Situação: {situacao}\n")
print("Dados salvos em aluno.txt!")
```

```
#lista inicial
convidados = ['João', 'Ana', 'Maria', 'Ana', 'Joaquim']

#removendo duplicatas
convidados = list(set(convidados))

#função de adicionar nome
def adicionar_nome(lista , nome):
  if nome not in lista:
    lista.append(nome)

#entrada do usuário
nome_digitado = input("Digite seu nome")
adicionar_nome(convidados, nome_digitado)

#verificação
if nome_digitado in convidados:
  print("Nome presente na lista!")
else:
  print("Nome NÃO encontrado!")

#loop para exibir todos
print("\nLista final de convidados:")
for pessoa in convidados:
  print("-", pessoa)

#Salvando um arquivo
with open("presenca.txt", "w") as arq:
  for pessoa in convidados:
    arq.write(pessoa + "\n")

print("\nLista salva em presenca.txt!")
```
