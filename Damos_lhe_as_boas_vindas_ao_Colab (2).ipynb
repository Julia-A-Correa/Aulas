**Dicionário (dict)** você **pode pensar neles como uma agenda de contatos, onde o nome é a chave, e o telefone o valor, ou seja, são compostos por chave: valor.**

```
Aluno = {'nome' : 'Ana' , 'idade' : '16', 'notas': [7.5, 8.0, 9.0]}
media = sum(Aluno['notas']) / len(Aluno['notas'])
matérias = {'matematica', 'português', 'inglês' , 'português'}
foi_aprovado = media >=7
print(f"nome: {Aluno['nome']}")
print(f"média:{media}")
print(f" matérias: {matérias}")
print(f"Aprovado? {foi_aprovado}")
```
